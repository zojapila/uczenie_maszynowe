```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
                
```

```python
# 1. Proszę wyświetlić dane i zapoznać się z posczególnymi informacjami.
iris
```

```python
# 2. Wyświetl liczbę wierszy oraz kolumn.
print('rows',len(iris.axes[0]))
print('columns', len(iris.axes[1]))
```

```python
# 3. Wyświetl podstawowe informacje dla poszczególnych kolumn. Skorzystaj z metod describe.
iris.describe()
```

```python
# 4. Zapoznaj się z działaniem metody groupby.
iris.groupby(['target']).mean()
```

```python
# 5. Wykorzystując metodę head proszę wyświelić 5 pierwszych wierszy
iris.head(5)
```

```python
# 6. Korzystając z dropna sprawdź, czy baza zawiera brakujące dane
iris.dropna()
```

```python
# 7. Korzystając z metody sort_values posortuj dane rosnąco względem drugiej kolumny.
iris.sort_values(['sepal width (cm)'])
```

```python
# 8. Wyznacz minimalną i maksymalną długość płatka (kolumna 3 - petal length). Podaj indeks tych wartości (min, idxmin).
print(' wartosć minimalna: ', iris['petal length (cm)'].min())
print('indeks: ',iris['petal length (cm)'].idxmin())
print(' wartosć maksymalna: ', iris['petal length (cm)'].max())
print('indeks: ',iris['petal length (cm)'].idxmax())

```

```python
# 9. Oblicz odchylenie standardowe dla każdej kolumny (std)
iris.std()
```

```python
# 10. Wyodrębnij wiersze dla których długość kielicha kwiatów (sepal length) jest większa od średniej długości (cały zbiór)
df = iris[iris['sepal length (cm)'] > iris['sepal length (cm)'].mean()]
df
```

```python
# 11. Wyświetl histogramy dla poszczególnych parametrów z uwzglednieniem przynależności do danej klasy
plt.hist([iris[iris['target'] == 0.0]['sepal length (cm)'], iris[iris['target'] == 1.0]['sepal length (cm)'], iris[iris['target'] == 2.0]['sepal length (cm)']])
plt.title('sepal length (cm)')
plt.legend(['0', '1', '2'])
```

```python
plt.hist([iris[iris['target'] == 0.0]['sepal width (cm)'], iris[iris['target'] == 1.0]['sepal width (cm)'], iris[iris['target'] == 2.0]['sepal width (cm)']])
plt.title('sepal width (cm)')
plt.legend(['0', '1', '2'])
```

```python
plt.hist([iris[iris['target'] == 0.0]['petal length (cm)'], iris[iris['target'] == 1.0]['petal length (cm)'], iris[iris['target'] == 2.0]['petal length (cm)']])
plt.title('petal length (cm)')
plt.legend(['0', '1', '2'])
```
